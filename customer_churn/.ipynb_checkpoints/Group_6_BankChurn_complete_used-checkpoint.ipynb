## Feature Engineering

Annual data set of the macro-economic database AMECO, used to replace 'geography' column so that this model can be used for other countries

https://ec.europa.eu/info/business-economy-euro/indicators-statistics/economic-databases/macro-economic-database-ameco/download-annual-data-set-macro-economic-database-ameco_en

```python
country_data = {'Geography': ['Germany', 'Spain', 'France'],
                'UnemploymentRate': [0.034, 0.153, 0.09],
                'ConsumerPriceIndex_b2015': [103.7756293, 103.4538192, 103.0908591],
                'GDP_perhead_euro': [40622.63, 25749.83, 35067.53]}
country_df = pd.DataFrame(data=country_data)
new_df = df.merge(country_df, left_on='Geography', right_on='Geography')
```

```python
new_df["BALANCE_TENURE"] = new_df.apply(lambda x: x["Balance"] / x["Tenure"] if x["Tenure"] != 0 else 0, axis = 1)
new_df["SALARY_BALANCE"] = new_df.apply(lambda x: x["EstimatedSalary"] / x["Balance"] if x["Balance"] != 0 else 0, axis = 1)
new_df['SCORE_AGE'] = new_df.apply(lambda x: x["CreditScore"] / x["Age"] if x["Age"] != 0 else 0, axis = 1)
```

```python
fig, axarr = plt.subplots(3, 2, figsize=(20, 12))
sns.histplot(new_df, x="BALANCE_TENURE", hue="Exited", element="step", stat="density", common_norm=False, ax=axarr[0][0])
sns.histplot(new_df, x="SALARY_BALANCE", hue="Exited", element="step", stat="density", common_norm=False, ax=axarr[0][1])
sns.histplot(new_df, x="SCORE_AGE", hue="Exited", element="step", stat="density", common_norm=False, ax=axarr[1][0])
sns.histplot(new_df, x="UnemploymentRate", hue="Exited", element="step", stat="density", common_norm=False, ax=axarr[1][1])
sns.histplot(new_df, x="ConsumerPriceIndex_b2015", hue="Exited", element="step", stat="density", common_norm=False, ax=axarr[2][0])
sns.histplot(new_df, x="GDP_perhead_euro", hue="Exited", element="step", stat="density", common_norm=False, ax=axarr[2][1])
```

## Feature Selection

Looking at 'Exited' column of correlation matrix and the EDA of categorical features, we can choose some correlated features first

```python
chosen_features_corr = ['Age', 'Balance',
                        'ConsumerPriceIndex_b2015', 'GDP_perhead_euro','CreditScore']
chosen_features_cat = ['HasCrCard', 'Gender', 'IsActiveMember']
label = ['Exited']

total_features = chosen_features_corr.copy()
total_features.extend(chosen_features_cat)

total_columns = total_features.copy()
total_columns.extend(label)
```

```python
model_df = new_df[total_columns]
model_df
```

```python
for col in chosen_features_cat:
    model_df = pd.get_dummies(model_df, columns=[col], drop_first=True)

total_features_encode = total_features.copy()
total_features_encode.remove('HasCrCard')
total_features_encode.remove("Gender")
total_features_encode.remove("IsActiveMember")
total_features_encode.extend(["HasCrCard_1", "Gender_Male", "IsActiveMember_1"])

model_df
```

## Modelling

```python
X_train, X_test, y_train, y_test = train_test_split(model_df[total_features_encode], model_df[label], 
                                                    test_size=0.20, stratify=model_df[label], random_state=2022)
```

```python
scaler = RobustScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Building Classification Model

```python
import matplotlib.pyplot as plt
def model_Evaluate(model):
    # accuracy of model on training data
    acc_train = model.score(X_train_scaled, y_train)
    # accuracy of model on test data
    acc_test = model.score(X_test_scaled, y_test)
    
    print('Accuracy of model on training data : {}'.format(acc_train*100))
    print('Accuracy of model on testing data : {} \n'.format(acc_test*100))

    # Predict values for Test dataset
    y_pred = model.predict(X_test_scaled)
    y_pred_proba = model.predict_proba(X_test_scaled)[::,1]
    
    # precision of model on test data
    pre_test = precision_score(y_test, y_pred)
    # recall of model on test data
    rec_test = recall_score(y_test, y_pred)
    # f1 of model on test data
    f1_test = f1_score(y_test, y_pred)
    # f2 of model on test data
    f2_test = fbeta_score(y_test, y_pred, beta=2)
    # AUC of model on test data
    auc_test = roc_auc_score(y_test, y_pred_proba)
    
    
    # Print the evaluation metrics for the dataset.
    print(classification_report(y_test, y_pred))
    
    # Compute and plot the Confusion matrix
    cf_matrix = confusion_matrix(y_test, y_pred)

    categories  = ['Negative','Positive']
    group_names = ['True Neg','False Pos', 'False Neg','True Pos']
    group_percentages = ['{0:.2%}'.format(value) for value in cf_matrix.flatten() / np.sum(cf_matrix)]

    labels = [f'{v1}\n{v2}' for v1, v2 in zip(group_names,group_percentages)]
    labels = np.asarray(labels).reshape(2,2)

    sns.heatmap(cf_matrix, annot = labels, cmap = 'Reds',fmt = '',
                xticklabels = categories, yticklabels = categories)

    plt.xlabel("Predicted values", fontdict = {'size':14}, labelpad = 10)
    plt.ylabel("Actual values"   , fontdict = {'size':14}, labelpad = 10)
    plt.title ("Confusion Matrix", fontdict = {'size':18}, pad = 20)
    
    d = {'Train_Accuracy': [acc_train], 'Test_Accuracy': [acc_test], 
         'Precision': [pre_test], 'Recall': [rec_test],
         'AUC': [auc_test], 'F1_Score': [f1_test], 'F2_Score': [f2_test]}
    
    df = pd.DataFrame(data=d)
    # summarize feature importance
    print(type(model).__name__)
    modelName = type(model).__name__
    if (modelName not in exclude_model):
        
        if(modelName in featureimportance) :
            impt = zip(X_train.columns, model.feature_importances_)
        else:
            impt = zip(X_train.columns, model.coef_[0])
        sorted_impt = sorted(impt, key = lambda x:x[1])
        for feat, importance in sorted_impt:
            print('feature: {f}, importance: {i}'.format(f=feat, i=importance))
        fig = plt.figure(figsize =(10, 7))
        plt.bar([item[0] for item in sorted_impt], [item[1] for item in sorted_impt], color ='maroon',width = 0.2)
        plt.show()
    return df
    
```

```python
featureimportance = ['XGBClassifier','RandomForestClassifier','KNeighborsClassifier']
exclude_model = ['CalibratedClassifierCV', 'KNeighborsClassifier','StackingClassifier']
```

```python
error_rate = []
acc_score = []
rec_score = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i, weights = 'uniform')
    knn.fit(X_train_scaled,y_train)
    pred_i = knn.predict(X_test_scaled)
    error_rate.append(np.mean(pred_i != y_test['Exited'].tolist()))
    acc_score.append(accuracy_score(y_test, pred_i))
    rec_score.append(recall_score(y_test, pred_i))

plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.plot(range(1,40),acc_score,color='green', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.plot(range(1,40),rec_score,color='orange', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate / Accuracy / recall vs. K Value')
plt.xlabel('K')
plt.ylabel('Rate')
print("Minimum error:-",min(error_rate),"at K =",error_rate.index(min(error_rate)))
print("Maximum accuracy:-",max(acc_score),"at K =",acc_score.index(max(acc_score)))
print("Maximum recall:-",max(rec_score),"at K =",rec_score.index(max(rec_score)))
```

# Ensemble Learning

```python
from sklearn.ensemble import StackingClassifier
```

```python
estimators = [
    ('lg', lg1),
    ('xgb', xgb_class),
    ('rf', rf1),
    ('nb', nb),
    ('knn', knn),
    ('svm', svm1),
]
stacked_clf = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression(class_weight='balanced'))

stacked_clf.fit(X_train_scaled,y_train)
stacked_clf_df = model_Evaluate(stacked_clf)
```
